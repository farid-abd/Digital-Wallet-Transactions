### Investigating the relationship between payment methods and transaction success rates


From data above, we can see that every payment method have high successful rate. But the next question is, what is the most successful payment method between the option

```python
crosstab = pd.crosstab(wallet_trans['payment_method'], wallet_trans['transaction_status'])

crosstab.plot(kind='bar', figsize=(10, 6), stacked=True)
plt.xlabel('Payment Method')
plt.ylabel('number_of_status')
plt.xticks(rotation=45)
plt.legend(title='Transaction Status')
plt.show()
crosstab
```

Bank transfer is the most successful payment method because it have the high number of successful transaction. Followed by debit card and then between credit card and UPI. Wallet balance is the last option people choose for paying their transaction. 

### Exploring seasonal trends in digital wallet usage

```python
wallet_trans[["transaction_date", "product_category", "product_amount"]]
```

```python
wallet_trans['transaction_date'] = pd.to_datetime(wallet_trans['transaction_date'])

# Extract components
wallet_trans['Year'] = wallet_trans['transaction_date'].dt.year
wallet_trans['Month'] = wallet_trans['transaction_date'].dt.month
wallet_trans['Day'] = wallet_trans['transaction_date'].dt.day
wallet_trans['Hour'] = wallet_trans['transaction_date'].dt.hour

wallet_trans
```

```python
trans_2023 = wallet_trans.loc[wallet_trans['Year'] < 2024, "product_category"]
trans_2023
```

### Analyzing spending patterns across different product categories

```python
product_spent = wallet_trans.groupby('product_category')['product_amount'].sum().reset_index()
product_spent = product_spent.sort_values(by='product_amount', ascending=True)
product_spent
```

```python
product_types = product_spent['product_category']
total_amount = product_spent['product_amount']
```

```python
product_types
```

```python
total_amount
```

```python
# Plot the pie chart
plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(
    total_amount, 
    labels=product_types, 
    autopct='%1.1f%%',  # Show percentages with one decimal place
    startangle=140,     # Start angle for the first slice
    colors=plt.cm.Pastel2.colors
)

# Add a title
plt.title("Product Amount by Category")

# Display the chart
plt.show()
```

The pie chart above displaying the percentage of every product in the total amoung of digital transactions. It is important tot note that the graph still displayed it in the total amount of each transaction. There's still transaction fee and cashback that not yet calculated.
